```python
import pandas as pd
baby_df = pd.read_csv('baby.csv')
baby_df.replace({True : 1,False: 0}, inplace =  True)
baby_df.rename(columns = {'Unnamed: 0' : 'id'}, inplace = True)
baby_df.drop('id', axis=1, inplace=True)
smoke = baby_df[baby_df['Maternal.Smoker'] == 1] 
nonsmoke = baby_df[baby_df['Maternal.Smoker'] == 0] 
smoke.head()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

plt.figure(figsize = (20,10))

ax = sns.heatmap(smoke.corr(),linewidths=.5 ,annot = True)

ax.set(xlabel='Indicators for Birthweight', ylabel='Indicators for Birthweight')

plt.show()

```
